```python
"""
Task Four – Search and Optimization
Breadth First Search (BFS) and Depth First Search (DFS)
Finds and displays the path from initial node to goal state
"""

from collections import deque
from typing import Dict, List, Set, Tuple, Optional

class GraphSearch:
    """
    A class to perform BFS and DFS graph searches
    """

    def __init__(self, graph: Dict[str, List[str]]):
        """
        Initialize the graph

        Args:
            graph: Adjacency list representation of the graph
                  Example: {'A': ['B', 'C'], 'B': ['A', 'D'], ...}
        """
        self.graph = graph

    def breadth_first_search(self, start: str, goal: str) -> Tuple[Optional[List[str]], int]:
        """
        Perform Breadth First Search to find path from start to goal

        Args:
            start: Initial node
            goal: Target node

        Returns:
            Tuple of (path from start to goal, number of nodes visited)
            Returns (None, visited_count) if no path exists
        """
        # Queue stores tuples of (current_node, path_to_current_node)
        queue = deque([(start, [start])])
        visited = set([start])
        visited_count = 0

        print("=" * 60)
        print(f"BFS SEARCH - Finding path from '{start}' to '{goal}'")
        print("=" * 60)

        while queue:
            current_node, path = queue.popleft()
            visited_count += 1

            print(f"Step {visited_count}: Visiting node '{current_node}'")
            print(f"  Current path: {' -> '.join(path)}")

            # Check if we reached the goal
            if current_node == goal:
                print(f"\n✓ GOAL REACHED! Path found after visiting {visited_count} nodes")
                print(f"  Final path: {' -> '.join(path)}")
                return path, visited_count

            # Explore neighbors
            for neighbor in self.graph.get(current_node, []):
                if neighbor not in visited:
                    visited.add(neighbor)
                    new_path = path + [neighbor]
                    queue.append((neighbor, new_path))
                    print(f"  Adding neighbor '{neighbor}' to queue (path: {' -> '.join(new_path)})")

        print(f"\n✗ No path found from '{start}' to '{goal}' after visiting {visited_count} nodes")
        return None, visited_count

    def depth_first_search(self, start: str, goal: str) -> Tuple[Optional[List[str]], int]:
        """
        Perform Depth First Search to find path from start to goal

        Args:
            start: Initial node
            goal: Target node

        Returns:
            Tuple of (path from start to goal, number of nodes visited)
            Returns (None, visited_count) if no path exists
        """
        # Stack stores tuples of (current_node, path_to_current_node)
        stack = [(start, [start])]
        visited = set()
        visited_count = 0

        print("=" * 60)
        print(f"DFS SEARCH - Finding path from '{start}' to '{goal}'")
        print("=" * 60)

        while stack:
            current_node, path = stack.pop()
            visited_count += 1

            print(f"Step {visited_count}: Visiting node '{current_node}'")
            print(f"  Current path: {' -> '.join(path)}")

            # Check if we reached the goal
            if current_node == goal:
                print(f"\n✓ GOAL REACHED! Path found after visiting {visited_count} nodes")
                print(f"  Final path: {' -> '.join(path)}")
                return path, visited_count

            # Mark as visited
            if current_node not in visited:
                visited.add(current_node)

                # Explore neighbors (push to stack in reverse order for natural DFS order)
                neighbors = self.graph.get(current_node, [])
                for neighbor in reversed(neighbors):
                    if neighbor not in visited:
                        new_path = path + [neighbor]
                        stack.append((neighbor, new_path))
                        print(f"  Adding neighbor '{neighbor}' to stack (path: {' -> '.join(new_path)})")

        print(f"\n✗ No path found from '{start}' to '{goal}' after visiting {visited_count} nodes")
        return None, visited_count

    def compare_searches(self, start: str, goal: str):
        """
        Compare BFS and DFS performance for the same search

        Args:
            start: Initial node
            goal: Target node
        """
        print("\n" + "=" * 60)
        print(f"COMPARISON: BFS vs DFS from '{start}' to '{goal}'")
        print("=" * 60)

        # Run BFS
        bfs_path, bfs_visited = self.breadth_first_search(start, goal)

        print("\n" + "-" * 60 + "\n")

        # Run DFS
        dfs_path, dfs_visited = self.depth_first_search(start, goal)

        # Display comparison
        print("\n" + "=" * 60)
        print("SEARCH COMPARISON SUMMARY")
        print("=" * 60)
        print(f"BFS - Path found: {bfs_path}")
        print(f"BFS - Path length: {len(bfs_path) if bfs_path else 'N/A'}")
        print(f"BFS - Nodes visited: {bfs_visited}")
        print(f"\nDFS - Path found: {dfs_path}")
        print(f"DFS - Path length: {len(dfs_path) if dfs_path else 'N/A'}")
        print(f"DFS - Nodes visited: {dfs_visited}")

        if bfs_path and dfs_path:
            print(f"\nComparison:")
            print(f"  BFS found shorter path? {len(bfs_path) <= len(dfs_path)}")
            print(f"  BFS visited fewer nodes? {bfs_visited <= dfs_visited}")


def create_sample_graph() -> Dict[str, List[str]]:
    """
    Create a sample graph for demonstration

    Graph structure:
        A --- B --- D
        |     |
        |     |
        C --- E --- F

    Returns:
        Dictionary representing the graph
    """
    graph = {
        'A': ['B', 'C'],
        'B': ['A', 'D', 'E'],
        'C': ['A', 'E'],
        'D': ['B'],
        'E': ['B', 'C', 'F'],
        'F': ['E']
    }
    return graph


def create_tree_graph() -> Dict[str, List[str]]:
    """
    Create a tree structure for demonstration

    Tree structure:
            A
           / \
          B   C
         / \   \
        D   E   F
           /
          G

    Returns:
        Dictionary representing the tree
    """
    graph = {
        'A': ['B', 'C'],
        'B': ['D', 'E'],
        'C': ['F'],
        'D': [],
        'E': ['G'],
        'F': [],
        'G': []
    }
    return graph


def main():
    """
    Main function to demonstrate BFS and DFS algorithms
    """
    print("\n" + "=" * 60)
    print("GRAPH SEARCH DEMONSTRATION - BFS vs DFS")
    print("=" * 60)

    # Example 1: Simple graph with cycles
    print("\n\n【 EXAMPLE 1: Simple Graph with Cycles 】")
    graph1 = create_sample_graph()
    print("\nGraph Structure:")
    for node, neighbors in graph1.items():
        print(f"  {node} -> {neighbors}")

    search1 = GraphSearch(graph1)
    search1.compare_searches('A', 'F')

    # Example 2: Tree structure (no cycles)
    print("\n\n【 EXAMPLE 2: Tree Structure (No Cycles) 】")
    graph2 = create_tree_graph()
    print("\nTree Structure:")
    for node, neighbors in graph2.items():
        print(f"  {node} -> {neighbors}")

    search2 = GraphSearch(graph2)
    search2.compare_searches('A', 'G')

    # Example 3: Custom graph demonstration
    print("\n\n【 EXAMPLE 3: Custom Graph - User Input 】")
    print("Let's create a simple custom graph:")
    print("Example: A connected to B and C; B connected to D; C connected to D")

    custom_graph = {
        'A': ['B', 'C'],
        'B': ['D'],
        'C': ['D'],
        'D': []
    }

    print("\nCustom Graph Structure:")
    for node, neighbors in custom_graph.items():
        print(f"  {node} -> {neighbors}")

    search3 = GraphSearch(custom_graph)

    # Interactive search
    print("\n--- Interactive Search ---")
    start_node = input("Enter start node (or press Enter for 'A'): ").strip().upper()
    goal_node = input("Enter goal node (or press Enter for 'D'): ").strip().upper()

    if not start_node:
        start_node = 'A'
    if not goal_node:
        goal_node = 'D'

    if start_node in custom_graph and goal_node in custom_graph:
        search3.compare_searches(start_node, goal_node)
    else:
        print(f"Error: Nodes '{start_node}' or '{goal_node}' not found in graph!")
        print(f"Available nodes: {list(custom_graph.keys())}")


if __name__ == "__main__":
    main()
```
